```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
import time

```

```python


def extract_youtube_comments(driver, number_of_scrolls=5):
    """
    Extract comments from a YouTube video page.
    
    Parameters:
    - driver: The Selenium WebDriver instance.
    - number_of_scrolls: The number of times to scroll down to load more comments (default is 5).
    
    Returns:
    - A list of extracted comments.
    """
    
    # Scroll down the page to load comments
    body = driver.find_element(By.TAG_NAME, "body")
    for _ in range(number_of_scrolls):
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(2)  #  some time to load comments
    
    # Wait for comments to be loaded
    try:
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, "//ytd-comment-thread-renderer"))
        )
    except Exception as e:
        print(f"Comments did not load in time: {e}")
        return []
    
    # Extract comments
    comments = driver.find_elements(By.XPATH, "//ytd-comment-thread-renderer")
    extracted_comments = []
    
    for comment in comments:
        try:
            text = comment.find_element(By.ID, "content-text").text
            extracted_comments.append(text)
        except Exception as e:
            print(f"Error extracting comment: {e}")
            continue
    
    return extracted_comments

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://www.youtube.com/watch?v=0jY4v4NDfcE")  

# Wait for the page to fully load
try:
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.TAG_NAME, "body"))
    )
except Exception as e:
    print(f"Page did not load in time: {e}")
    driver.quit()
    exit()

# Adding additional sleep to ensure all elements are loaded
time.sleep(10)

comments_list = extract_youtube_comments(driver, number_of_scrolls=5)

# Print comments with indices 
for index, comment in enumerate(comments_list, start=1):
    print(f"{index}. {comment}")

# Save comments to a CSV file
csv_file = "youtube_comments.csv"
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Comment"])
    for comment in comments_list:
        writer.writerow([comment])

print(f"Comments have been saved to {csv_file}")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get('https://www.amazon.sg/SAMSUNG-Android-Lasting-Included-Speakers/product-reviews/B09WZ13CBS/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews')

```

```python
from selenium.webdriver.common.by import By
import time
import pandas as pd

def extract_and_next(driver, next_button_xpath):
    cus_elements = driver.find_elements(By.CSS_SELECTOR, "div.a-profile-content > span.a-profile-name")

    review_elements = driver.find_elements(By.CSS_SELECTOR, "span[data-hook='review-body'] > span")


    cus_list = [item.text for item in cus_elements]
    reviews_list = [item.text for item in review_elements]
    
    try:
        next_button = driver.find_element(By.XPATH, next_button_xpath)
        next_button.click()
        time.sleep(2)
    except:
        print(f"Error on page {driver.current_url}")
    
    return cus_list, reviews_list

next_button_xpath = '//a[contains(@href, "/product-reviews/") and contains(text(), "Next page")]'
data = []

num_pages = 10
for _ in range(num_pages):
    names, reviews = extract_and_next(driver, next_button_xpath)
    for name, review in zip(names, reviews):
        data.append([name, review])

df = pd.DataFrame(data, columns=['CustomerName', 'Review'])
df.to_csv('reviews2.csv', index=False)
# df.to_excel('reviews.xlsx', index=False)  # Save in Excel format
```
